# Data Cleaning

Based on the above analysis we can see that a lot of our columns need cleaning as some numerical columns are object type because of alphabets or , attached in it.

1. Size Column should be numeric and hence needs cleaning to change its dtype.
2. Released and Last Updated Columns should be Datetime and not object type.
3. Content Rating Column needs to be cleaned by merging few similar categories.
4. Minimum Android needs a lot of cleaning as there are values with versions 1.1.1 type, 1.1 and up type and 1.1-2.1 type., hence need to bring into one common value type and also change the dtype.
5. Currency Column needs to be treated as well. If wherever the app is paid, currency is different then it has to be brought in same currency.
6. Free column and Price column might show multicollinearity if wherever free is true, price is 0, we will drop the Free Column.
7. After Missing Value treatment, we need to clean Category column as well. We need to merge few non performing categories into others category and rest top 15-20 best performing categories.

```python
model_data['Size']=np.where(model_data['Size']=='Varies with device', np.nan, model_data['Size'])
```

```python
def impute_size(size):
    size=str(size).replace(',', '')
    if size[-1]=='M':
        size=size[:-1]
        return float(size)
    elif size[-1]=='k':
        size=size[:-1]
        return float(size)/1024
    elif size[-1]=='G':
        size=size[:-1]
        return float(size)*1000
    else:
        return np.nan
```

```python
model_data['Size']=model_data['Size'].apply(impute_size)
```

**4. Minimum Android**

```python
model_data['Minimum Android'].sample(10)
```

**There are multiple levels of cleaning involved in this column.**

1. There are values with 4.0 and up type.
2. There are values like 4.0.3
3. There are values like 4.3w and up
4. There are values in range like 2.1-3.1
5. There are values like Varies with device

**Solution/Approach**
1. Remove and up 
2. As per version syntax, the last number in x.y.z version of a product generally denotes a bug fixes, so we can ignore that.
3. We need to remove w as these are syntactical to versions only.
4. We need to create a separate column where we could keep the upper range of the value as Maximum Android and minimum range as minimum android. For those which we do not have maximum android, we will take the max version present in the data.
5. We will simply treat these as nan and impute accordingly.

```python
model_data[['Minimum Android', 'Maximum Android']]=model_data['Minimum Android'].str.split('-', expand=True)
```

```python
model_data['Minimum Android']=model_data['Minimum Android'].apply(lambda x: str(x)[:3])
```

```python
model_data['Maximum Android']=model_data['Maximum Android'].apply(lambda x: str(x)[:3])
```

```python
# So while removing the '-' part, we also removed e from none and made it a value so let's fix this
model_data['Minimum Android']=np.where(model_data['Minimum Android']=='nan', np.nan, model_data['Minimum Android'])
```

```python
def impute_installs(max_installs):
    
    if max_installs == 0:
        return '0+'
    elif 1 <= max_installs < 5:
        return '1+'
        
    elif 5 <= max_installs < 10:
        return '5+'

    elif 10 <= max_installs < 50:
        return '10+'
    elif 50 <= max_installs < 100:
        return '50+'
        
    elif 100 <= max_installs < 500:
        return '100+'
       
    elif 500 <= max_installs < 1000:
        return '500+'
    elif 1000 <= max_installs < 5000:
        return '1,000+'
    elif 5000 <= max_installs < 10000:
        return '5,000+'
        
    elif 10000 <= max_installs < 50000:
        return '10,000+'
    elif 50000 <= max_installs < 100000:
        return '50,000+'
    elif 100000 <= max_installs < 500000:
        return '100,000+'
    elif 500000 <= max_installs < 1000000:
        return '500,000+'
    elif 1000000 <= max_installs < 5000000:
        return '1,000,000+'
        
    elif 5000000 <= max_installs < 10000000:
        return '5,000,000+'
    elif 10000000 <= max_installs < 50000000:
        return '10,000,000+'
    elif 50000000 <= max_installs < 100000000:
        return '50,000,000+'
    elif 100000000 <= max_installs < 500000000:
        return '100,000,000+'
    elif 500000000 <= max_installs < 1000000000:
        return '500,000,000+'
    else:
        return '1,000,000,000+'
    
```

```python
model_data['Installs'].fillna(model_data['Maximum Installs'].apply(impute_installs), inplace=True)
```

**Minimum Installs**

```python
def impute_min_installs(max_installs):
    
    if max_installs == 0:
        return 0
    elif 1 <= max_installs < 5:
        return 1
        
    elif 5 <= max_installs < 10:
        return 5

    elif 10 <= max_installs < 50:
        return 10
    elif 50 <= max_installs < 100:
        return 50
        
    elif 100 <= max_installs < 500:
        return 100
       
    elif 500 <= max_installs < 1000:
        return 500
    elif 1000 <= max_installs < 5000:
        return 1000
    elif 5000 <= max_installs < 10000:
        return 5000
        
    elif 10000 <= max_installs < 50000:
        return 10000
    elif 50000 <= max_installs < 100000:
        return 50000
    elif 100000 <= max_installs < 500000:
        return 100000
    elif 500000 <= max_installs < 1000000:
        return 500000
    elif 1000000 <= max_installs < 5000000:
        return 1000000
        
    elif 5000000 <= max_installs < 10000000:
        return 5000000
    elif 10000000 <= max_installs < 50000000:
        return 10000000
    elif 50000000 <= max_installs < 100000000:
        return 50000000
    elif 100000000 <= max_installs < 500000000:
        return 100000000
    elif 500000000 <= max_installs < 1000000000:
        return 500000000
    else:
        return 1000000000
    
```

```python
model_data['Minimum Installs'].fillna(model_data['Maximum Installs'].apply(impute_min_installs), inplace=True)
```
